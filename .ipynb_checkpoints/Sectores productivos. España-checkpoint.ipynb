```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup
```

```python
dfsectores=pd.read_excel(".xlsx")
```

```python
df=pd.read_excel("1_comex_sectores.xlsx")
```

```python
valor_a_multiplicar = (-1)
df['Importaciones'] = df['Importaciones'] * valor_a_multiplicar
df1=df.head(23)
df2=df.tail(46)
df3=df2.head(23)
df4=df2.tail(23)
```

```python
min_value = min(df1[['Importaciones', 'Exportaciones']].min())
max_value = max(df1[['Importaciones', 'Exportaciones']].max())

# Configurar el gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Crear barras para las exportaciones
ax.bar(df1['Sectores Económicos'], df1['Exportaciones'], color='red', alpha=0.5, label='Exportaciones')

# Crear barras para las importaciones, con valor cero en el centro del eje Y
importaciones_negativas = df1['Importaciones'].apply(lambda x: min(0, x))
importaciones_positivas = df1['Importaciones'].apply(lambda x: max(0, x))
ax.bar(df1['Sectores Económicos'], importaciones_positivas, color='blue', alpha=0.5, label='Importaciones')
ax.bar(df1['Sectores Económicos'], importaciones_negativas, color='blue', alpha=0.5)

# Crear línea para la balanza comercial
balanza_comercial = df1['Exportaciones'] + df1['Importaciones']
ax.plot(df1['Sectores Económicos'], balanza_comercial, color='green', marker='o', label='Balanza Comercial')

# Ajustar límites del eje Y
plt.ylim(-max_value * 1.1, max_value * 1.1)

# Agregar etiquetas y leyenda
ax.set_title('Importaciones, Exportaciones y Balanza Comercial por Sector Económico')
ax.set_xlabel('Sectores Económicos')
ax.set_ylabel('Valor')
ax.axhline(0, color='black', linewidth=0.5)  # Línea en cero
ax.legend()

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Por los datos ya presentados vemos que la mayor carencia que tiene España se centra en la importación de recursos que se transforman en energía (ya sea gasolina y otros combustibles como elecctricidad). Por ello voy a centrarme en analizar cómo se organiza la producción energética en España y si es posible la evolución del consumo de combustibles fósiles en España de cara a la futura implantación de coches eléctricos.


Analicemos cómo ha ido evolucionando el sistema eléctrico en España a través de su consumo y producción

El siguiente gráfico se usará para ver únicamente cuanto se produce y se consume, pero lo interesante de aquí es saber el porcentaje de cada una de las fuentes de energía.

```python
rel = requests.get('https://datosmacro.expansion.com/energia-y-medio-ambiente/electricidad-consumo/espana')
htmlel = rel.content
soupel = BeautifulSoup(htmlel, 'html.parser')
soupel.prettify
soupel.find_all('table', attrs={'class': 'table tabledat table-striped table-condensed table-hover'})
html_tableel = soupel.find_all('table', attrs={'class': 'table tabledat table-striped table-condensed table-hover'})[0]
dfelectricidad = pd.read_html(html_tableel.prettify())[0]
```

```python
y1 = dfelectricidad['Generación GWh']
y2 = dfelectricidad['Consumo GWh'] 
x=dfelectricidad['Fecha']
plt.plot(x,y1,marker='d',linestyle='--', color='black', label='Generación GWh')
plt.legend()
plt.xlabel('Fecha')
plt.ylabel('GWh')
plt.title('Relación entre Producción y Consumo de energía')
plt.bar(x,y2, label='Consumo GWh')
plt.figure(figsize=(12, 8))
plt.show()
```
